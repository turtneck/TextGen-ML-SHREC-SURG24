```python
import pandas as pd,os,re,sys
cnt=0;end=10
filename = 'D:\\prompt\\3_5M-GPT3_5-Augmented_edit-256-1.csv'
df_iter = pd.read_csv(filename, iterator=True, chunksize=1)
while True:
    try:
        if not end is None: 
            if cnt>=end:break
        df = next(df_iter)
        
        print( str( list(df.question)[0]) )
        print("-----")
        print( str( list(df.response)[0]) )
        print("==============================")
        cnt+=1
    except StopIteration:
        break
```
